# **Data Visualization of Indian Startups by State**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('indian_startups_by_state.csv')
```

# Indian Startups by State :

```python
plt.figure(figsize=(25, 10))
sns.countplot(data=df, x='State', palette='viridis')
plt.title('Indian Startups by State', fontweight='bold')
plt.xlabel('State', fontweight='bold')
plt.ylabel('Count', fontweight='bold')
plt.xticks(fontweight='bold', rotation=90)
plt.tight_layout()

plt.show()
```

# Trends in Startup Growth :

```python
startup_count_by_year = df.groupby('Year')['Count'].sum()
```

```python
plt.figure(figsize=(15, 10)) 
startup_count_by_year.plot(kind='line', marker='o', linestyle='--', markersize=8) 
plt.ylabel('Number of Startups')
plt.title('Trends in Startup Growth')
plt.show()
```

# Distribution of Startups by State :

This bar chart illustrates the number of startups in each state, showcasing the distribution of startups across India. The states are sorted from largest to smallest based on the number of startups.

```python
startup_count_by_state = df.groupby('State')['Count'].sum()
```

```python
startup_count_by_state = startup_count_by_state.sort_values(ascending=False)
```

```python
plt.figure(figsize=(25, 10))
startup_count_by_state.plot(kind='bar', color='skyblue')
plt.xlabel('State')
plt.ylabel('Number of Startups')
plt.title('Distribution of Startups by State')
plt.xticks(rotation=90)
plt.show()
```

# Distribution of Startups by Industry :

This bar chart showcases the distribution of startups across different industries in India. It provides insights into the distribution of startups by industry and highlights the diversity of the startup ecosystem in the country. The industries are displayed on the x-axis, while the number of startups is represented on the y-axis. The chart allows for a comparison of the number of startups across different industries. The states are not specifically mentioned in this visualization, as it focuses solely on the distribution by industry.

```python
startup_count_by_industry = df.groupby('Industry')['Count'].sum()
```

```python
startup_count_by_industry = startup_count_by_industry.sort_values(ascending=False)
```

```python
plt.figure(figsize=(25, 10))
startup_count_by_industry.plot(kind='bar', color='orange')
plt.xlabel('Industry')
plt.ylabel('Number of Startups')
plt.title('Distribution of Startups by Industry')
plt.xticks(rotation=90)
plt.show()
```
